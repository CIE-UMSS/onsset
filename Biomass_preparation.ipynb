```python
# Import necessary modules and libraries

import pandas as pd
from pandas import DataFrame as df
import os
import tkinter as tk
from tkinter import filedialog, messagebox
```

```python
#open the input (csv) file
root = tk.Tk()
root.withdraw()
root.attributes("-topmost", True)
messagebox.showinfo('BIOMASS', 'Open the input file with extracted GIS data ')
input_file = filedialog.askopenfilename()
#path = r"C:\GHANA"
#name_of_file = "GHANA_Updated.csv"
```

```python
# Import csv as pandas dataframe
input = pd.read_csv(input_file)
country ='Somalia'
```

```python
input.fillna(0,inplace=True)
```

```python
input_subset=df(input,columns= ['acof', 'bana', 'barl', 'bean', 'cass', 'chic', 'cnut', 'coco', 'cott',
       'cow_p', 'grou', 'lent', 'maiz', 'ocer', 'ofib', 'oilp', 'ooil', 'opul',
       'orts', 'pige', 'plnt', 'pmil', 'pota', 'rape', 'rcof', 'rest', 'rice',
       'sesa', 'smil', 'sorg', 'soyb', 'sugb', 'sugc', 'sunf', 'swpo', 'teas',
       'temf', 'toba', 'trof', 'vege', 'whea', 'yams'])
input_subset.head()
```

```python
#residue-product-ratio (RPR)
#suggestion:maybe make this column as a seperate excel sheet and import as a dataframe
#REFERENCES:
    #Khatiwada, D., Purohit, P., Ackom, E.K., 2019. Mapping Bioenergy Supply and Demand in Selected Least Developed Countries (LDCs): Exploratory Assessment of Modern Bioenergyâ€™s Contribution to SDG7. Sustainability 11, e7091. https://doi.org/Khatiwada D, Purohit P <http://pure.iiasa.ac.at/view/iiasa/239.html> ORCID: https://orcid.org/0000-0002-7265-6960 <https://orcid.org/0000-0002-7265-6960>, & Ackom EK  (2019).  Mapping Bioenergy Supply and Demand in Selected Least Developed Countries (LDCs): Exploratory Assessment of Modern Bioenergyâ€™s Contribution to SDG7.   Sustainability 11 (24): e7091. DOI:10.3390/su11247091 <https://doi.org/10.3390/su11247091>.
    #Okello, C., Pindozzi, S., Faugno, S., Lorenzo, B., 2013. Bioenergy potential of agricultural and forest residues in Uganda. Biomass Bioenergy 56. https://doi.org/10.1016/j.biombioe.2013.06.003
    #http://www.fao.org/3/AD576E/ad576e00.pdf
#assumptions:
#beans,lentils,peas=pulses ; othercereals==barley
#other fibre=jute ; otheroilcrops==palm oil ;other roots,yam ==potatoes
#plantain==banana ;rapeseed,sunflower==groudnut; restofcrops=2; rapeseed==groudnut

acof=1
bana=2
barl=1.3
bean=1.4
cass=0.88
chic=1.4
cnut=1.6
#coco
cott=1.1
cow_p=2.9
grou=2.1
lent=1.4
maiz=2.5
ocer= 1.3
ofib=2
oilp=0.24
ooil=0.24
opul=1.4
orts=0.25
pige=5
plnt=2
pmil=1.4
pota=0.25
rape= 2.1
rcof=1
rest=2
rice=1.8
sesa=2
smil=1.2
sorg=1.4
soyb=2.66
sugb=0.7
sugc=0.4
sunf=2.1
swpo=0.25
#teas=
temf=2
toba=0.2
trof=2
vege=1
whea=1.6
yams=0.25
```

```python
#compute the residuals in tonnes. The unit is stated here:https://www.mapspam.info/methodology/
input_subset['acof']=input_subset['acof']*acof
input_subset['bana']=input_subset['bana']*bana
input_subset['barl']=input_subset['barl']*barl
input_subset['bean']=input_subset['bean']*bean
input_subset['cass']=input_subset['cass']*cass
input_subset['chic']=input_subset['chic']*chic
input_subset['cnut']=input_subset['cnut']*cnut
input_subset['cott']=input_subset['cott']*cott
input_subset['cow_p']=input_subset['cow_p']*cow_p
input_subset['grou']=input_subset['grou']*grou
input_subset['lent']=input_subset['lent']*lent
input_subset['maiz']=input_subset['maiz']*maiz
input_subset['ocer']=input_subset['ocer']*ocer
input_subset['ofib']=input_subset['ofib']*ofib
input_subset['oilp']=input_subset['oilp']*oilp
input_subset['ooil']=input_subset['ooil']*ooil
input_subset['opul']=input_subset['opul']*opul
input_subset['orts']=input_subset['orts']*orts
input_subset['pige']=input_subset['pige']*pige
input_subset['plnt']=input_subset['plnt']*plnt
input_subset['pmil']=input_subset['pmil']*pmil
input_subset['pota']=input_subset['pota']*pota
input_subset['rape']=input_subset['rape']*rape
input_subset['rcof']=input_subset['rcof']*rcof
input_subset['rest']=input_subset['rest']*rest
input_subset['rice']=input_subset['rice']*rice
input_subset['sesa']=input_subset['sesa']*sesa
input_subset['smil']=input_subset['smil']*smil
input_subset['sorg']=input_subset['sorg']*sorg
input_subset['soyb']=input_subset['soyb']*soyb
input_subset['sugb']=input_subset['sugb']*sugb
input_subset['sugc']=input_subset['sugc']*sugc
input_subset['sunf']=input_subset['sunf']*sunf
input_subset['swpo']=input_subset['swpo']*swpo
input_subset['temf']=input_subset['temf']*temf
input_subset['toba']=input_subset['toba']*toba
input_subset['trof']=input_subset['trof']*trof
input_subset['vege']=input_subset['vege']*vege
input_subset['whea']=input_subset['whea']*whea
input_subset['yams']=input_subset['yams']*yams
```

```python
#adding the technical potential in tonnes 
technical_potential=input_subset['technical_potential']=input_subset.sum(axis=1)
print("The total bioenergy technical potential in {} is {:.3f} million tonnes".format(country,(technical_potential.sum(axis=0))/1000000))
```

```python
#calculation for bioelectricty potential
annual_full_load_hours =5600    #unit is hours
specific_biomass=0.00121           #unit is ton/kWh
specific_bagasse=0.0016            #unit is ton/kWh
elec_potential_coefficient=0.2  #assumption for agricultiral residues available for electricity production in the least developing countries
capacity_factor_biomass=0.8
capacity_factor_bagasse=0.53
biomass_subset=df(input_subset,columns= ['acof', 'bana', 'barl', 'bean', 'cass', 'chic', 'cnut', 'coco', 'cott',
       'cow_p', 'grou', 'lent', 'maiz', 'ocer', 'ofib', 'oilp', 'ooil', 'opul',
       'orts', 'pige', 'plnt', 'pmil', 'pota', 'rape', 'rcof', 'rest', 'rice',
       'sesa', 'smil', 'sorg', 'soyb', 'sunf', 'swpo', 'teas',
       'temf', 'toba', 'trof', 'vege', 'whea', 'yams'])
#bagasse_subset=df(input_subset,columns=['sugb', 'sugc']) --this was not used
biomass_subset[:] *=(elec_potential_coefficient/specific_biomass)
bioelec_potential_kWh=biomass_subset['bioelec_potential_kWh']=biomass_subset.sum(axis=1)
print("The total bioelectricity potential in {} is {:.3f} TWh".format(country,(bioelec_potential_kWh.sum(axis=0))/1000000000))
```

```python
#adding the bioelectricy_potential column to the original dataset and dropping the infomation of the individual criops
input=df(input,columns=['Country','NightLights', 'Pop', 'id', 'GridCellArea', 'ElecPop', 'WindVel', 'GHI',
       'TravelHours', 'Elevation', 'Slope', 'ResidentialDemandTIerCustom',
       'LandCover', 'SubstationDist', 'CurrentHVLineDist', 'CurrentMVLineDist',
       'RoadDist', 'X_deg', 'Y_deg', 'TransformerDist', 'PlannedMVLineDist',
       'PlannedHVLineDist', 'HydropowerDist', 'Hydropower', 'HydropowerFID','bioelec_potential_kWh'])
input['bioelec_potential_kWh']=biomass_subset['bioelec_potential_kWh']

input.head()
```
